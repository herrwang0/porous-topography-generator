```python
import cartopy
import matplotlib.pyplot as plt
import netCDF4
import numpy as np
```

```python
# Model mesh
with netCDF4.Dataset('OM5_ocean_hgrid.nc') as nc:
    Mlon = nc.variables['x'][::2,::2]
    Mlat = nc.variables['y'][::2,::2]
```

```python
with netCDF4.Dataset('new_topo_OM5_grid_r7_6x387.nc') as nc:
    R7topo = nc.variables['depth'][:]
    R7h2 = nc.variables['depth'][:]
R7h2[ R7topo<=0 ] = 0
R7topo[ R7topo<=0 ] = 0
```

```python
with netCDF4.Dataset('new_topo_OM5_grid_r6_3x129.nc') as nc:
    R6topo = nc.variables['depth'][:]
    R6h2 = nc.variables['depth'][:]
R6h2[ R6topo<=0 ] = 0
R6topo[ R6topo<=0 ] = 0
```

```python
with netCDF4.Dataset('new_topo_OM5_grid_r5_3x43.nc') as nc:
    R5topo = nc.variables['depth'][:]
    R5h2 = nc.variables['depth'][:]
R5h2[ R5topo<=0 ] = 0
R5topo[ R5topo<=0 ] = 0
```

```python
with netCDF4.Dataset('new_topo_OM5_grid_r3_2x9.nc') as nc:
    R3topo = nc.variables['depth'][:]
    R3h2 = nc.variables['depth'][:]
R3h2[ R3topo<=0 ] = 0
R3topo[ R3topo<=0 ] = 0
```

```python
def sq(A, pos=False):
    B = 1*A
    if pos:
        B[A<=0] = np.nan
    else:
        B[A==0] = np.nan
    return B
```

```python
plt.pcolormesh( Mlon, Mlat, sq(R7topo, pos=True) ); plt.colorbar();
plt.xlabel('Longitude ($^\circ$E)'); plt.xlabel('Latitude ($^\circ$N)');
plt.title('OM5 R7 depth (m)');
```

```python
plt.pcolormesh( Mlon, Mlat, sq(R6topo) ); plt.colorbar();
plt.title('OM5 R6 depth (m)');
```

```python
plt.pcolormesh( Mlon, Mlat, sq(R7topo - R6topo) ); plt.colorbar();
plt.xlabel('Longitude ($^\circ$E)'); plt.xlabel('Latitude ($^\circ$N)');
plt.title('OM5 R7 - R6 depth (m)');
```

```python
plt.pcolormesh( Mlon, Mlat, sq(R7topo - R5topo), cmap=plt.cm.RdYlBu_r , vmin=-3, vmax=3); plt.colorbar();
plt.xlabel('Longitude ($^\circ$E)'); plt.xlabel('Latitude ($^\circ$N)');
plt.title('OM5 R7 - R5 depth (m)');
```

```python
plt.pcolormesh( Mlon, Mlat, sq(R7topo - R5topo), cmap=plt.cm.RdYlBu_r , vmin=-3, vmax=3); plt.colorbar();
plt.xlim(-105,-65); plt.ylim(5,35);
plt.xlabel('Longitude ($^\circ$E)'); plt.xlabel('Latitude ($^\circ$N)');
plt.title('OM5 R7 - R5 depth (m)');
```

```python
plt.pcolormesh( Mlon, Mlat, sq(R7topo - R3topo), cmap=plt.cm.RdYlBu_r , vmin=-5, vmax=5); plt.colorbar();
plt.xlabel('Longitude ($^\circ$E)'); plt.xlabel('Latitude ($^\circ$N)');
plt.title('OM5 R7 - R3 depth (m)');
```
